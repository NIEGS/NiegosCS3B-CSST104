1. Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

2. Read the Data

```python
data = pd.read_csv('Lesson_4_student_grades_dataset.csv')
print(data)
```

3. Visualize the Data

```python
plt.scatter(data['Attendance_Percentage'],data['Travel_Time_to_School'])
plt.xlabel('Independent Variable(Travel_Time_to_School')
plt.xlabel('Dependent Variable (Attendance_Percentage)')
plt.title('Data')
plt.show()
```

4. Preparing Data for Modeling:

```python
X = data[['Attendance_Percentage']]
Y = data[['Travel_Time_to_School']]
```

5. Fitting the Linear Regression Model:

```python
model = LinearRegression()
model.fit(X,Y)
```

6. Getting Model Coefficients:

```python
slope = model.coef_[0]
intercept = model.intercept_
print("Slope:", slope)
print("Intercept:", intercept)
```

7. Visualizing the Linear Regression Line:

```python
plt.scatter(X,Y)
plt.plot(X,model.predict(X),color='red')
plt.xlabel('Independent Variable (Travel_Time_to_School)')
plt.ylabel('Dependent Variable (Attendance_Percentage)')
plt.title('Linear Regression')
plt.show()
```

8. Making Predictions


```python
new_X = [[np.random.randint(100,200)]]
predicted_Y = model.predict(new_X)
print("Value of X:", new_X)
print("Predicted Y for new X:", predicted_Y)
```
