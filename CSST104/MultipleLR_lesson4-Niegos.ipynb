# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

# Loading Dataset

```python
data = pd.read_csv('Lesson_4_student_grades_dataset.csv')
print(data)
```

# Visualizing the Dataset

```python
x = data.drop('Attendance_Percentage', axis=1)
y = data[['Travel_Time_to_School']]

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)

data.head()
```

# Initializing the Model

```python
model = LinearRegression()
```

# Fitting the Model

```python
model.fit(x_train,y_train)
```

# Making Predictions

```python
y_predict = model.predict(x_test)
print(x_test)
print(y_predict)
```

# Calculating the Mean Squared error

```python
mse = mean_squared_error(y_test,y_predict)
print(f"Mean Squared error: {mse}")
```

# Prediction

```python
Travel_Time_to_School = [[4, 63, 0, 24, 31]]
predicted_Travel_Time_to_School = model.predict(Travel_Time_to_School)
print(f"Predicted: {predicted_Travel_Time_to_School[0]}")
```
